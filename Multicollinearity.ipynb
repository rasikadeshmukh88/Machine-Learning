# Multicollinearity
Multicollinearity means that two or more independent (predictor) variables in a regression model are highly correlated with each other.

In simple words â€” when one predictor can be predicted from another, they contain redundant information.

**Example**
Suppose you are predicting house price using:

Size (in square feet)

Number_of_rooms

Since a larger house usually has more rooms, these two predictors are highly correlated â†’ this is multicollinearity.



**Why Itâ€™s a Problem**

* Makes regression coefficients unstable (large changes with small data variations).

* Makes it hard to interpret which variable actually affects the target.

* When predictors are highly correlated, it becomes difficult to determine their individual effect on the dependent variable because they carry overlapping information.

# How to Check for Multicollinearity

**1.Correlation Matrix**

Compute pairwise correlation between independent variables.

High correlation (|r| > 0.8) indicates possible multicollinearity.

**2.Variance Inflation Factor (VIF)**

VIF quantifies how much a variable is inflated due to correlation with others.

ð‘‰ð¼ð¹=1/1âˆ’ð‘…^2

A VIF > 5 or 10 indicates high multicollinearity.

**Remedies**

Remove or combine correlated features.

Use Principal Component Analysis (PCA).

Use regularization methods (Ridge or Lasso regression).

```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor

# --- Step 1: Create a dataset ---
np.random.seed(10)
X1 = np.random.rand(100)
X2 = 2 * X1 + np.random.rand(100) * 0.1   # Highly correlated with X1
X3 = np.random.rand(100)                   # Independent feature

df = pd.DataFrame({
    'X1': X1,
    'X2': X2,
    'X3': X3
})

# --- Step 2: Correlation matrix ---
print("Correlation Matrix:")
print(df.corr(), "\n")

# --- Step 3: Calculate VIF for each variable ---
vif_data = pd.DataFrame()
vif_data["Feature"] = df.columns
vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]

print("Variance Inflation Factor (VIF):")
print(vif_data)
```
