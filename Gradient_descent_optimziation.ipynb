```python
# Gradient descent for linear regression
# yhat = m*x + c
# loss = (y - yhat)^2/N
# to find these where we can get minimum loss
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create an random array
x = np.random.randn(10,1)
y = 2*x + np.random.rand()
```

```python
x
```

```python
# paramters
m = 0
c = 0
# hyper parameter
learning_rate = 0.01
```

```python
def decent(x,y,m,c,learning_rate):
  dldm = 0
  dldc = 0
  N = x.shape[0]
  # iterate loss function
  for xi,yi in zip(x,y):
    dldm = dldm + (-2*xi*(yi - (m*xi +c)))
    dldc = dldc + (-2*(yi-(m*xi +c)))
  m = m - learning_rate*(1/N)* dldm
  c = c - learning_rate *(1/N)* dldc

  return m, c


for epoch in range(400):
  m, c = decent(x,y,m,c,learning_rate)
  yhat = m *x + c
  loss = np.divide(np.sum((y - yhat) ** 2), x.shape[0])
  print(f'{epoch}: loss is {loss}, parameter m is {m} and c is {c}')
  plt.plot(x, yhat)
```
